7. What are the key properties of a Bernoulli distribution?
- Binary Outcomes: Bernoulli Distribution models situations with only two possible outcomes, such as success (1) and failure (0).
- Constant Probability: The probability of success (p) remains consistent across all trials in a Bernoulli Distribution.
- Independent Trials: The outcome of each trial is not influenced by subsequent trials.
- Complementary Probability: The probability of failure (q) can be found by subtracting the probability of success (p) from 1; q = 1-p.
- Discrete Probability Distribution: Bernoulli Distribution involves a finite number of different outcomes that make it a discrete distribution.
- Expected Value (Mean): The mean of a Bernoulli Distribution is equal to the probability of success (p), representing the average outcome.
- Variance: Variance (pq) measures how outcomes deviate from the mean in a Bernoulli Distribution.

- The normal distribution, also called the Gaussian distribution,
is one of the most important probability distributions in statistics. It describes a bell-shaped curve that is symmetric about the mean.

- The standard normal distribution is a special case of the normal distribution with:

It is a bell-shaped, symmetric probability distribution that describes how data values are distributed around the mean. It is denoted by the variable Z, and often called the Z-distribution.

Importance

Foundation of Many Statistical Models:

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem is a fundamental concept in statistics that states:

When independent random variables are added, their normalized sum tends toward a normal distribution  even if the original variables themselves are not normally distributed  as the sample size becomes large.

Enables Use of Normal Distribution:

Lets us use normal distribution techniques even when the population isn't normal.

Foundation of Inferential Statistics:

Used in creating confidence intervals and performing hypothesis testing.

Applies to Many Real-World Problems:

Many natural and business processes involve averages; CLT helps analyze them statistically.

Justifies Use of Z-tests and T-tests:

These tests assume that sample means follow a normal distribution, which CLT confirms for large n.

Simplifies Complex Distributions:

Instead of dealing with complicated or unknown distributions, we can rely on the normal distribution for averages.



14. How does the Central Limit Theorem relate to the normal distribution?

CLT creates a bridge to use the normal distribution for analyzing means from any population.

It justifies the use of normal distribution for many real-world problems, especially when working with averages or sums.

Even if the population is not normal, the distribution of sample means becomes normal  that's why the normal distribution is so widely used in statistics.

15. What is the application of Z statistics in hypothesis testing?

One-sample Z-test:
Test whether the sample mean significantly differs from a known population mean.

Example: Is the average battery life different from the claimed 10 hours?

Two-sample Z-test:
Compare means of two independent samples when variances are known and samples are large.

Example: Do two teaching methods result in different average scores?

Proportion Z-test:
Test if a sample proportion differs from a population proportion.

Example: Does the percentage of voters favoring a candidate differ from 50%?

Confidence Intervals:
Z-scores are used to construct confidence intervals for means or proportions when population standard deviation is known.

17. What are point estimates and interval estimates in statistics?

- Point Estimate:

A point estimate is a single value that serves as a best guess for an unknown population parameter.

Example:If a sample of 50 students has an average height of 160 cm, then:

Point estimate for population mean height = 160 cm

- Interval Estimate:

An interval estimate provides a range of values (interval) that is likely to contain the true population parameter, with a certain level of confidence.

Example:
If a 95% confidence interval for average height is (157 cm, 163 cm), we are 95% confident the true population mean lies within that range.

18. What is the significance of confidence intervals in statistical analysis?

 Significance of Confidence Intervals (CIs):

1.Estimates with Uncertainty
CIs give a range, not just a point.

This range accounts for sampling variability.

Example: Saying "The mean is 70  3" is more informative than just "mean is 70".

2.Helps in Decision Making
If a 95% CI for drug effectiveness is (75%, 90%), we can be confident the drug works well.

If a CI for a difference between groups includes 0, it suggests no significant difference.

3.lternative to Hypothesis Testing
CIs provide the same information as a test of significance, but with more context.

Instead of just saying significant or not, CIs tell you how large or small the effect could be.

4.Demonstrates Precision
A narrow CI means a more precise estimate.

A wider CI shows more uncertainty or variability in the data.

5.Supports Confidence Levels
Common confidence levels: 90%, 95%, 99%.

A 95% CI means if we repeated the study 100 times, about 95 of those intervals would contain the true value.

6.Interpretable for Non-statisticians

CIs make statistical results more intuitive and communicative, especially in reports and presentations.



19. What is the relationship between a Z-score and a confidence interval?

The Z-score determines the width of a confidence interval when the population standard deviation is known and the data follows (or approximates) a normal distribution.

The Z-score sets the "margin of error" in the confidence interval.

A higher confidence level  larger Z-score  wider interval (more conservative estimate).

A lower confidence level  smaller Z-score  narrower interval (less certainty).

20.  How are Z-scores used to compare different distributions?

Different Scales:
Z-scores convert different datasets to a common scale (standard normal distribution with mean 0, SD 1).

Fair Comparison:
You can compare values that originally come from different units or ranges, like:

SAT scores vs. ACT scores

Heights of men vs. heights of women

Test scores from different classes

Outlier Detection:
Z-scores help identify which data point is more unusual or extreme in its distribution.

21. What are the assumptions for applying the Central Limit Theorem?

- Random Sampling
The samples must be randomly selected from the population.

Ensures that the sample accurately represents the population.

- Independence of Observations

Each observation in the sample must be independent of the others.

No value should influence another.

Often satisfied when sampling with replacement, or when sampling without replacement from a large population (n < 10% of population).

- Sample Size (n) Should Be Large

For non-normal populations, a sample size of at least 30 is usually sufficient.

For heavily skewed or non-symmetric distributions, an even larger sample may be needed.

If the population is normal, CLT holds even with small samples.

22. What is the concept of expected value in a probability distribution?

The expected value (also called mean or expectation) of a probability distribution is the long-run average outcome of a random variable if an experiment is repeated many times.
